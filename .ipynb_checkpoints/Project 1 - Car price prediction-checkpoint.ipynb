```python
cars=pd.read_csv('https://raw.githubusercontent.com/akjadon/Finalprojects_DS/master/Car_pricing_prediction/CarPrice_Assignment.csv')
```

```python
# data modifying
cars['carcomp']=cars['CarName'].apply(lambda x: x.split(" ")[0])
```

```python
cars['carcomp']
```

```python
cars['carcomp'].value_counts()
```

```python
cars.loc[(cars['carcomp']=='vw') | (cars['carcomp']=='vokswagen'),'carcomp']='volkswagen'
cars.loc[cars['carcomp']=='Volkswagen', 'carcomp']='volkswagen'
```

```python
cars.loc[(cars['carcomp']=='porcshce'),'carcomp']='porsche'
```

```python
cars.loc[cars['carcomp']=='toyouta','carcomp']='toyota'
```

```python
cars.loc[cars['carcomp']=='Nissan', 'carcomp']='nissan'
```

```python
cars.loc[cars['carcomp']=='maxda','carcomp']='mazda'
```

```python
cars['carcomp'].value_counts()
```

```python
cars.info()
```

```python
cars.head()
```

```python
cars = cars.drop('CarName', axis=1)
```

```python
cars.info()
```

```python
#2. Data preparation
x=cars.iloc[:,1:-2]
```

```python
x['carcomp']=cars['carcomp']
```

```python
x
```

```python
y=cars['price']
```

```python
y.head()
```

```python
#type change
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
carscat=x.select_dtypes(include=['object'])
```

```python
carscat.head()
```

```python
carscatchange=carscat[:].apply(le.fit_transform)
```

```python
carscatchange
```

```python
carsnum=x.select_dtypes(include=['float64','int64'])
```

```python
X=pd.concat([carsnum,carscatchange], axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
#3. Linear regression model building and testing
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

```python
y2=lr.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y2,y_test)
```

```python
metrics.r2_score(y2,y_test)
```

```python
#OLS model
import statsmodels.api as sm
```

```python
model=sm.OLS(y_train,x_train).fit()
```

```python
print(model.summary())
```

```python
y_pred = model.predict(sm.add_constant(x_test))
```

```python
fig,ax = plt.subplots()
ax.scatter(y_pred, y_test)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
fig.show()
```

```python
metrics.r2_score(y_test, y_pred)
```
